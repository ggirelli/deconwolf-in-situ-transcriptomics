# Figure 4

```r
plist = list(
    A=readRDS("panels_rds/fig_4a.rds"),
    B=readRDS("panels_rds/fig_4b.rds"),
    C=readRDS("panels_rds/fig_4c.rds"),
    D=readRDS("panels_rds/fig_4d.rds")
)
```

```r
options(repr.plot.width=18, repr.plot.height=12)
p = plot_grid(
    plist[[1]],
    plot_grid(plotlist=plist[2:4], ncol=3, align="hv", axis="blt", labels=letters[2:4]),
    nrow=2, labels=letters[1]
)
p
```

```r
ggsave(plot=p, file="figures/fig_4.png", width=15, height=10, bg="white")
ggsave(plot=p, file="figures/fig_4.jpg", width=15, height=10, bg="white")
```
